# Data Understanding 

At the beginning, the most important step of a machine learning project is a first analysis of the data. The aim of this notebook is to gain insights into the data and visualize them. 

## Import Statements

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# import data 
data = pd.read_csv("data.csv")

# get overview of features, data types & null values
print(data.info())

# correlation heatmap of numeric features
numeric_data = data.drop(['postal_code', 'id'], axis=1)
corr = numeric_data.corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)

```

## Visualization of category and sub-category

```python
fig, axes = plt.subplots(1, 2)

axes[0].set_ylabel('')
axes[0].set_title('Category', fontsize=16)
axes[1].set_ylabel('')
axes[1].set_title('Sub-Category', fontsize=16)

data['category'].value_counts().plot(kind='bar', ax=axes[0], fontsize=14)
data['sub_category'].value_counts().plot(kind='bar', ax=axes[1], figsize=(12,7), fontsize=14)

fig.tight_layout()
plt.show()
```

## Visualization of quantity and discount

```python
fig, axes = plt.subplots(1, 2)

axes[0].set_ylabel('', fontsize=16)
axes[0].set_title('Quantity', fontsize=18)
axes[1].set_ylabel('', fontsize=16)
axes[1].set_title('Discount', fontsize=18)

data['quantity'].plot(kind='box', ax=axes[0], fontsize=16)
data['discount'].plot(kind='box', ax=axes[1], figsize=(8,5), fontsize=16)

fig.tight_layout()
plt.show()
```

## Visualization of sales and profit

```python
fig, axes = plt.subplots(1, 2)

axes[0].set_ylabel('US-Dollar', fontsize=16)
axes[0].set_title('Sales', fontsize=18)
axes[1].set_ylabel('US-Dollar', fontsize=16)
axes[1].set_title('Profit', fontsize=18)

data['sales'].plot(kind='box', ax=axes[0], fontsize=16)
data['profit'].plot(kind='box', ax=axes[1], figsize=(8,10), fontsize=16)

fig.tight_layout()
plt.show()
```
